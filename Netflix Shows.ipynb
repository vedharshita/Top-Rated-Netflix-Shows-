```
# !/usr/bin/env python
# coding: utf-8
#Loading Libraries

from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```
data = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/GIT/Netflix shows/Netflix Shows.csv",encoding = 'ISO-8859-1')
data.head()
```

```
vec = CountVectorizer(lowercase=True, vocabulary=['action','rude','violence','humor','fantasy','adult','scary','parental','parents','mature'], decode_error='ignore')
```

```
data = data.dropna(subset=['ratingLevel']).reset_index()
```

```
x = vec.fit_transform(data['ratingLevel'])
```

```
df1 = pd.DataFrame(x.todense(), columns = ['action','rude','violence','humor','fantasy','adult','scary','parental','parents','mature'])
```

```
df1.to_csv("/content/gdrive/My Drive/Colab Notebooks/GIT/Netflix shows/NetflixData_genre.csv")
data.to_csv("/content/gdrive/My Drive/Colab Notebooks/GIT/Netflix shows/NetflixData_mod.csv")
```
